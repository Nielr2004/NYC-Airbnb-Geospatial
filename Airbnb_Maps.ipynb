```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster, HeatMap
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alexeygrigorev/datasets/master/AB_NYC_2019.csv"
df = pd.read_csv(url)

df = df.dropna(subset=['latitude', 'longitude', 'price', 'number_of_reviews'])

manhattan_data = df[df['neighbourhood_group'] == 'Manhattan'].head(1000)

print(f"Data Loaded: {len(manhattan_data)} listings ready to map.")
df.head(3)
```

```python
m = folium.Map(location=[40.730610, -73.935242], zoom_start=12, tiles='CartoDB dark_matter')

marker_cluster = MarkerCluster().add_to(m)

for idx, row in manhattan_data.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"<b>{row['name']}</b><br>${row['price']}/night",
        icon=folium.Icon(color='blue', icon='home')
    ).add_to(marker_cluster)


m.save("map.html")
```

```python
m2 = folium.Map(location=[40.730610, -73.935242], zoom_start=12)

heat_data = [[row['latitude'], row['longitude'], row['price']] for index, row in manhattan_data.iterrows()]

HeatMap(heat_data, radius=15).add_to(m2)

m2.save("map2.html")
```

```python
gems = df[(df['price'] < 150) & (df['number_of_reviews'] > 100) & (df['neighbourhood_group'] == 'Manhattan')]

m3 = folium.Map(location=[40.730610, -73.935242], zoom_start=13)

for idx, row in gems.head(50).iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        tooltip=f"{row['name']} (${row['price']})",
        icon=folium.Icon(color='green', icon='star')
    ).add_to(m3)

print(f"Found {len(gems)} hidden gems!")
m3.save("map3.html")
```
